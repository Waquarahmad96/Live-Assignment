```python
import numpy as np

# Create a NumPy array with random integers between 1 and 10 of shape (5, 6)
array = np.random.randint(1, 11, size=(5, 6))
print("Original array:")
print(array)
```

```python
# Extract all even integers
even_integers = array[array % 2 == 0]
print("Even integers:")
print(even_integers)
```

```python
# Extract all odd integers
odd_integers = array[array % 2 != 0]
print("Odd integers:")
print(odd_integers)
```

```python
import numpy as np

# Set the seed for reproducibility
np.random.seed(0)

# Create a 3x3x3 array with random integers between 1 and 10
array = np.random.randint(1, 11, size=(3, 3, 3))
print("Original array:")
print(array)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create two different Series, each of length 50
series1 = pd.Series(np.random.rand(50))
series2 = pd.Series(np.random.rand(50))

# Display the first few elements of each Series
print("Series 1:")
print(series1.head())
print("\nSeries 2:")
print(series2.head())

# Step 2: Plotting the Series using Matplotlib

# Create a figure
plt.figure(figsize=(14, 6))

# Plot series1
plt.subplot(1, 2, 1)
plt.plot(series1, color='blue', label='Series 1')
plt.title('Line Plot of Series 1')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()

# Plot series2
plt.subplot(1, 2, 2)
plt.plot(series2, color='green', label='Series 2')
plt.title('Line Plot of Series 2')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()

# Step 3: Plotting histograms of the Series

# Create a new figure for histograms
plt.figure(figsize=(14, 6))

# Histogram of series1
plt.subplot(1, 2, 1)
plt.hist(series1, bins=10, color='blue', alpha=0.7, label='Series 1')
plt.title('Histogram of Series 1')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Histogram of series2
plt.subplot(1, 2, 2)
plt.hist(series2, bins=10, color='green', alpha=0.7, label='Series 2')
plt.title('Histogram of Series 2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create the first Series with random numbers ranging from 10 to 50
series1 = pd.Series(np.random.uniform(10, 50, 50))

# Step 2: Create the second Series with random numbers ranging from 100 to 1000
series2 = pd.Series(np.random.uniform(100, 1000, 50))

# Step 3: Create a DataFrame by joining these Series by column
df = pd.DataFrame({
    'col1': series1,
    'col2': series2
})

# Display the DataFrame
print("DataFrame:")
print(df.head())

# Step 4: Plotting the Series from the DataFrame using Matplotlib

# Create a figure
plt.figure(figsize=(14, 6))

# Plot col1
plt.subplot(1, 2, 1)
plt.plot(df['col1'], color='blue', label='col1')
plt.title('Line Plot of col1')
plt.xlabel('Index')
plt.ylabel('Values (10 to 50)')
plt.legend()

# Plot col2
plt.subplot(1, 2, 2)
plt.plot(df['col2'], color='green', label='col2')
plt.title('Line Plot of col2')
plt.xlabel('Index')
plt.ylabel('Values (100 to 1000)')
plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()

# Step 5: Plotting histograms of the columns from the DataFrame

# Create a new figure for histograms
plt.figure(figsize=(14, 6))

# Histogram of col1
plt.subplot(1, 2, 1)
plt.hist(df['col1'], bins=10, color='blue', alpha=0.7, label='col1')
plt.title('Histogram of col1')
plt.xlabel('Value (10 to 50)')
plt.ylabel('Frequency')
plt.legend()

# Histogram of col2
plt.subplot(1, 2, 2)
plt.hist(df['col2'], bins=10, color='green', alpha=0.7, label='col2')
plt.title('Histogram of col2')
plt.xlabel('Value (100 to 1000)')
plt.ylabel('Frequency')
plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
